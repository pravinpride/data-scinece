```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
level = np.array([1,2,3,4,5,6])
```

```python
level
```

```python
salary = np.array([100,50,200,250,60,150])
```

```python
salary
```

```python
plt.bar(level,salary)
plt.title('Test')
plt.xlabel('Level')
plt.ylabel('SALARY')
plt.show()
```

```python
plt.bar(level,salary)
plt.title('Test',fontsize=25)
plt.xlabel('Level',fontsize=25)
plt.ylabel('SALARY',fontsize=25)
plt.show()
```

```python
plt.bar(level,salary,color='red')
plt.title('Test',fontsize=25)
plt.xlabel('Level',fontsize=25)
plt.ylabel('SALARY',fontsize=25)
plt.show()
```

```python
plt.scatter(level,salary)
plt.title('Test',fontsize=25)
plt.xlabel('Level',fontsize=25)
plt.ylabel('SALARY',fontsize=25)
plt.show()
```

```python
plt.plot(level,salary)
plt.title('Test',fontsize=25)
plt.xlabel('Level',fontsize=25)
plt.ylabel('SALARY',fontsize=25)
plt.show()
```

```python
plt.plot(level,salary,marker='o',markersize=15,markerfacecolor='red',ls='--',lw=8)
plt.title('Test',fontsize=25)
plt.xlabel('Level',fontsize=25)
plt.ylabel('SALARY',fontsize=25)
plt.show()
```

```python
plt.pie(salary)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter(level,salary)
plt.subplot(1,2,2)
plt.bar(level,salary)
plt.show()
```

```python
plt.subplot(1,3,1)
plt.scatter(level,salary)
plt.subplot(1,3,2)
plt.bar(level,salary)
plt.subplot(1,3,3)
plt.pie(salary)
plt.tight_layout()
plt.show()
```

```python
plt.subplot(3,1,1)
plt.scatter(level,salary)
plt.subplot(3,1,2)
plt.bar(level,salary)
plt.subplot(3,1,3)
plt.pie(salary)
plt.tight_layout()
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(level,salary)
axes.set_title('ISHA')
axes.set_xlabel('LEVEL')
axes.set_ylabel('SALARY')
plt.show()
```

```python
incentives = np.array([56,2,9,20,60,25])
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.4,0.3,0.4,0.4])
axes1.plot(level,salary)
axes2.plot(level,incentives)
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(level,salary)
axes[1].plot(level,incentives)
plt.tight_layout()
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(level,salary,label='level vs salary')
axes.plot(level,incentives,label='level vs incen')
plt.legend(loc=0)
plt.show()
```
